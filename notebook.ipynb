```python
def analizar_lexico(input_str):
    # Definicion del AFD
    transiciones = {
        'q0': [
            (lambda c: c == '$', 'q1'),
            (lambda c: c.isdigit(), 'q3'),
        ],
        'q1': [
            (lambda c: c.isalnum(), 'q2'),
        ],
        'q2': [
            (lambda c: c.isalnum(), 'q2'),
        ],
        'q3': [
            (lambda c: c.isdigit(), 'q3'),
            (lambda c: c == '.', 'q4'),
        ],
        'q4': [
            (lambda c: c.isdigit(), 'q5'),
        ],
        'q5': [
            (lambda c: c.isdigit(), 'q5'),
        ]
    }

    # Aqui van los codigos para cada token
    estados_aceptacion = {
        'q2': 100,
        'q3': 150,
        'q5': 200
    }

    # Aqui estan relacionados los codigos con los token
    codigos = {
        100: 'identificador',
        150: 'entero',
        200: 'decimal'
    }

    index = 0
    resultados = []
    longitud = len(input_str)

    while index < longitud:
        estado_actual = 'q0'
        inicio = index
        ultimo_aceptado = None
        pos_aceptado = -1

        # Procesar hasta encontrar un token valido o error
        while index < longitud:
            char = input_str[index]
            transicion_encontrada = False

            # Verificar transiciones para el estado actual
            for condicion, estado_destino in transiciones.get(estado_actual, []):
                if condicion(char):
                    estado_actual = estado_destino
                    transicion_encontrada = True
                    break

            if not transicion_encontrada:
                break  # No hay transición válida

            # Verificar si el nuevo estado es de aceptación
            if estado_actual in estados_aceptacion:
                ultimo_aceptado = estado_actual
                pos_aceptado = index

            index += 1

        # Manejar resultados después de procesar
        if ultimo_aceptado is not None:
            # Token reconocido
            token = input_str[inicio:pos_aceptado + 1]
            codigo = estados_aceptacion[ultimo_aceptado]
            resultados.append(f"{token} = {codigos[codigo]}")
            index = pos_aceptado + 1
        else:
            # Manejar errores
            if index == inicio:
                # No hubo ninguna transición desde q0
                resultados.append(f"{input_str[inicio]} = token no existe")
                index += 1
            else:
                # Backtracking: token no reconocido
                token = input_str[inicio:index]
                resultados.append(f"{token} = token no reconocido")

    return resultados

# Prueba con el caso de ejemplo
# entrada = "$name.200.0"
entrada = "$$$hola ."
resultados = analizar_lexico(entrada)

for res in resultados:
    print(res)
```
